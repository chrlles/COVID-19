## COVID-19 na América do Sul nos ultímos 7 dias

 Análise de casos e óbitos de COVID-19 na América do Sul na ultímos 7 dias
 <br>fonte dados: https://www.kaggle.com/anandhuh/covid-weekly-trends-south-america

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv("covid_south_america_weekly_trend.csv")
dados.head()
```

Casos nos ultímos 7 dias

```python
fig, ax = plt.subplots(figsize=(7,9))
plt.barh(dados['Country/Other'], dados['Cases in the last 7 days'])
plt.style.use('ggplot')
plt.xlabel('Número de Casos nos ultímos 7 dias')
plt.ylabel('País')
plt.title('Número de Casos em cada País nos ultímos 7 dias')
```

Obitos nos ultímos 7 dias

```python
fig, ax = plt.subplots(figsize=(7,9))
plt.barh(dados['Country/Other'], dados['Deaths in the last 7 days'], color='#81A649')
ax.tick_params(labelsize=11)
plt.style.use('seaborn-bright')
plt.ylabel('País')
plt.xlabel('número de Óbitos no ultímos 7 dias')
plt.title('Número de Óbitos na América de Sul')
```

Os gráficos de casos e obitos são bastante parecidos, mas o números de Casos está em proporção maior que o gráfico de Óbitos.

Em ambos o Brasil está como destaque, com grande número de Casos e Óbitos.
